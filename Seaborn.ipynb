Dist Plot -> sns.distplot(single variable)

Joint Plot -> sns.jointplot(first variable, second variable, dataset)

Pair Plot -> sns.pairplot(dataset)

Rug Plot -> Similar to distplot                    sns.rugplot(single variable)

Bar Plot -> sns.barplot(categorical variable, numerical variable, dataset)

Count Plot -> sns.countplot(categorical variable, dataset)

Box Plot -> sns.boxplot(categorical variable, numerical variable, dataset)

Violin Plot -> sns.violinplot(categorical variable, numerical variable, dataset)

Swarm Plot -> sns.swarmplot(categorical variable, numerical variable, dataset)

CatPlot -> sns.catplot(categorical variable, numerical variable, dataset, kind)

HeatMap -> sns.heatmap(data which is the correlation or the pivot table)

Cluster Map -> sns.clustermap(data which is the correlation or the pivot table)

# Dist Plot
Distribution of uni-variate or one variable set of observation

```python
sns.distplot(tips['total_bill']) # Displays a Histogram and a Kernel Density Estimatation
```

```python
# If you want to remove KDE

sns.distplot(tips['total_bill'], kde = False, bins = 20)  # Bins increase the no.of points
```

# Joint Plot

```python
sns.jointplot(x = 'total_bill', y = 'tip' , data = tips)    # Used to make Dist Plot for bi-variate distribution

# Scatter Plot + Histogram for 2 variables
```

```python
sns.jointplot(x = 'total_bill', y = 'tip' , data = tips, kind = 'hex')  # Default is scatter (Scatter Plot)

# If the hexagon with more points gets darker and light for less points
```

```python
sns.jointplot(x = 'total_bill', y = 'tip' , data = tips, kind = 'reg')  

# Draws a regression line for the scatter plot and the histogram
```

```python
sns.jointplot(x = 'total_bill', y = 'tip' , data = tips, kind = 'kde')   # Shows here the points match up the most
```

# Pair Plot

Basically plots pair wise relationship over the entire dataframe

Pair plot basically plots a join plot for each possible combination of the columns in the dataframe.

```python
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')   # Same column over the each other, it plots a histogram

# Hue only works for categorical columns
# Palette will have built in strings to set the color and style
```

# Rug Plot

```python
sns.rugplot(tips['total_bill'])

# Draws a dash line for each of the item for the dataset. 
# It draws one line over the other if the same point in encountered.
```

# Kernel Density Estimation

```python
sns.kdeplot(tips['total_bill'])
```

# Categorical Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

Distribution of the categorical plot

# Bar Plot

```python
sns.barplot(x = 'sex' , y = 'total_bill' , data = tips) # X -> Categorical Column y -> Numerical 
```

```python
import numpy as np
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.std )  # Default estimator is the mean 
```

# Count Plot

```python
sns.countplot(x = 'sex', data = tips)  # Same as Bar Plot but counts the no.of the values in the category
```

Shows the distribution of categorical data
Box Plot -> Box and Whisker Plot

# Box Plot

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips)

# Whiskers are the other points in the distribution
# Dots are the outliers
```

```python
# Set the hue as a categorical data

sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')
```

# Violin Plot

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips , hue = 'sex') 

# Shows more information -> Entire dataset
```

```python
# To set both the categories in the same plot, use split = True

sns.violinplot(x = 'day', y = 'total_bill', data = tips , hue = 'sex', split = True) 
```

# Strip Plot

Draws a scatter plot where one variable is categorical

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips)
```

```python
# You can't really tell how many points are stacked on top of each other. So use jitter = True 
# Split has been renamed to dodge 

sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = True, hue = 'sex', dodge = True)
```

# Swarm Plot 

Very similar to Strip Plot except the points are adjusted such that they are not overlapped

Disadvantages -> Does not scale these points to very large datasets.

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips)
```

```python
# Combine Swarm Plot and Violin Plot
sns.swarmplot(x = 'day', y = 'total_bill', data = tips)
sns.violinplot(x = 'day', y = 'total_bill', data = tips, color = 'black')
```

# CatPlot -> General Form of all these plots using the kind parameter

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'factor')
```

# Matrix Plot

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
flights.head()
```

# Heat Map

For a heat map to work, your data should be in a matrix form. Index and column value match. 
All the data in the table must have a index and a column name. For example: The index column does not have a column name. 

# Correlation 

```python
tc = tips.corr()
```

```python
tc
```

```python
sns.heatmap(data = tc, annot = True) # Annot basically represents the number into each box
```

```python
sns.heatmap(data = tc, annot = True, cmap = 'coolwarm')
```

# Pivot Table

```python
fp = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
```

```python
fp
```

```python
sns.heatmap(data = fp, cmap = 'magma', linewidth = 2, linecolor = 'white')  # magma is default
```

# Cluster Map

Uses hierachical clustering and clusters the rows and columns based on its similarity

```python
sns.clustermap(data = fp, cmap = 'coolwarm')
```

To standardize the scales

```python
sns.clustermap(data = fp, cmap = 'coolwarm', standard_scale= 1)
```

# Grids

```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
# To find the unique values in a list

iris['species'].unique()
```

```python
sns.pairplot(iris)
```

# Pair Grid

Gives you more control. We can plot our desired plots either in the lower, upper or diagonal section

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

# Facet Grid

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```

# Regression Plot

LMPlot -> Can represent linear models using Seaborn

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x = 'total_bill', y= 'tip', data = tips, hue = 'sex', markers = ['o','v'], scatter_kws ={'s':100})  # You can pass matplotlib parameters

# It calls matplotlib using lmplot
# scatter_kws -> You use to customize the markers using a dictionary.
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'sex', row = 'time', hue = 'smoker')  # Basically calls Facet Grids 
```

# Styling and Color

```python
sns.set_style('ticks')   # white, ticks,darkgrid,whitegrid
sns.countplot(x = 'sex', data = tips)
sns.despine()   # It will remove the top and the right grid ticks
```

```python
# If you want to remove the lines in the left and the bottom, use it as a parameter
sns.set_style('ticks')   # white, ticks,darkgrid,whitegrid
sns.countplot(x = 'sex', data = tips)
sns.despine(left = True, bottom = True)
```

# Size and Aspect

```python
plt.figure(figsize = (12,3)) # It overrrides the seaborn styles using the matplotlib figures.
sns.countplot(x = 'sex', data = tips)
```
