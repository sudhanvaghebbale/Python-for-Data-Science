# 2 Ways of creating plot -> Functional Method and Object Oriented

# Functional Method

```python
plt.plot(x,y,'r') # r -> Denotes 'Red' as the line color
```

plt.show() -> Basically prints out the graph instead of giving it as the output!

```python
# Functional

plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

# Multiplots on the same canvas

```python
plt.subplot(1,2,1) # plt.subplot(row,columns,plot number)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])  # Left, Bottom, Width, Height

axes.plot(x,y)

axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # Insert Axes

axes1.set_title('Larger Plot')
axes2.set_title('Smaller Plot')

axes1.plot(x,y)
axes2.plot(y,x)
```

Subplots using object oriented method

```python
fig, axes = plt.subplots(nrows = 3, ncols = 3) # Tuple Unpacking

#axes.plot(x,y)

plt.tight_layout() # Gives space between each plot
```

```python
axes
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

for cur in axes:
    cur.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

# Figure Size and DPI

```python
fig = plt.figure(figsize = (8,2), dpi = 100 )  # figsize is width and height of figure in inches
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (5,2))  # figsize is width and height of figure in inches
plt.tight_layout()

axes[0].plot(x,y)
axes[1].plot(y,x)
```

# Legends

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,label = 'X Squared')
axes.plot(x,x**3,label = 'X Cubed')

axes.legend()  # If you don't give label in the plot function, an warning will occur stating no handles to reference
# To make sure the legend does not overlap with the plot, use loc = 0
```

# Customizating Plots

# Plot Appearance -> Setting Colors

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y, color = '#FF0000') # color can take common colors(blue,red) or rgb hex code (#FF8C00)
```

# Line Width and Line Style

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

# instead of linewidth, you can use lw = 10
axes.plot(x,y, linewidth = 10, alpha = 0.5) # 1 is the default width  # alpha is amount of transparency
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

# instead of linestyle, you can use ls = '--'
axes.plot(x,y, linewidth = 5, linestyle= '--')
```

# Markers

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

# instead of linestyle, you can use ls = '--'
axes.plot(x,y, marker = 'o', markersize = 10, markerfacecolor = 'red', markeredgecolor = 'black', markeredgewidth = 2)
```

# Plot Range

```python
# To range only point on the x-axis from 0 to 1

fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y, lw = 2, ls = '--')

axes.set_xlim([0,1])
axes.set_ylim([0,2])
```
